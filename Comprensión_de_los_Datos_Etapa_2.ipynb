# Base de Datos encuesta AF 2 más completa para comparar

```python
# Base de Datos
import pandas as pd

df = pd.read_excel("2. Base Datos encuesta AF 2_mas completa para comparar.xlsx", sheet_name="BASE FINAL")
```

```python
# Dimensión del dataset
print(df.shape)
print(df.columns)
```

```python
# Comprensión de las variables
df.info()
```

```python
# Tabla descriptiva con variables de interés
variables = ['Edad madre', 'Educación', 'Región','Sexo hijo','PN hijo (g)', 'EG hijo (sem)','Condición mamá', 'Patología RN', '¿Consume SAF?', 'mgAF/día SAF', 'mgAF/día Suplementos y multivitamínico 1°T','Total mg/d AF  suple y pan', '¿Consumió pan?', 'mgAF/unidad pan' ]

df[variables].describe()
```

```python
df[variables].info()
```

```python
# Variables cuantitativas
# Medidas de tendencia central: promedio, mediana y moda de los datos.
medidas_data = {
    'Variable': [],
    'Media': [],
    'Mediana': [],
    'Moda': []
}

for col in variables:
    medidas_data['Variable'].append(col)
    medidas_data['Media'].append(df[col].mean())
    medidas_data['Mediana'].append(df[col].median())
    medidas_data['Moda'].append(df[col].mode()[0])

# Crear un DataFrame a partir de diccionario
cuantitativas_df = pd.DataFrame(medidas_data)

# Mostrar el DataFrame
display(cuantitativas_df)
```

```python
# Medidas de dispersión: rango: máximo - mínimo, varianza, desviación estándar
disp_data = {
    'Variable': [],
    'Rango': [],
    'Varianza': [],
    'Desviación Estándar': []
}

for col in variables:
    disp_data['Variable'].append(col)
    disp_data['Rango'].append(df[col].max() - df[col].min())
    disp_data['Varianza'].append(df[col].var())
    disp_data['Desviación Estándar'].append(df[col].std())

dispersion_df = pd.DataFrame(disp_data)
display(dispersion_df)
```

```python
# Variables cuantitativas:
# Medidas de posición no-central: cuartiles, outlier (valores atípicos), boxplots

import matplotlib.pyplot as plt

# Seleccionar solo las cuantitativas
vars_cuant = df[variables].select_dtypes(include = ['int64', 'float64']).columns
excluir = ['Educación', 'Región', 'Sexo hijo', 'Condición mamá',
           'Patología RN', '¿Consume SAF?', '¿Consumió pan?']

# CUartiles
cuartiles_df = pd.DataFrame({
    'Q1': df[vars_cuant].quantile(0.25),
    'Q2 (Mediana)': df[vars_cuant].quantile(0.5),
    'Q3': df[vars_cuant].quantile(0.75)
})

print("\nCuartiles:")
display(cuartiles_df)

# Outliers
resumen_outliers = []
detalle_outliers = []

for col in vars_cuant:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    inferior = Q1 - 1.5 * IQR
    superior = Q3 + 1.5 * IQR

    mask = (df[col] < inferior) | (df[col] > superior)

    # Resumen
    resumen_outliers.append({
        'Variable': col,
        'Q1': Q1,
        'Q3': Q3,
        'IQR': IQR,
        'Límite inferior': inferior,
        'Límite superior': superior,
        '# Outliers': mask.sum(),
        '% Outliers': mask.mean() * 100
    })

    # Separación bajos vs altos
    detalle_outliers.append({
        'Variable': col,
        'Outliers bajos': (df[col] < inferior).sum(),
        'Outliers altos': (df[col] > superior).sum()
    })

resumen_outliers_df = pd.DataFrame(resumen_outliers)
detalle_outliers_df = pd.DataFrame(detalle_outliers)

print("\nResumen de Outliers:")
display(resumen_outliers_df)

print("\nDetalle de Outliers (bajos vs altos):")
display(detalle_outliers_df)

# Boxplots
for col in vars_cuant:
    plt.figure()
    df.boxplot(column=col)
    plt.title(f'Boxplot de {col}')
    plt.grid()
    plt.show()
```

```python
#Análisis de distribución de los datos (Histogramas). Identificar si tiene forma simétrica o asimétrica
#Cálculo del coeficiente de sesgo
asimetria = df[vars_cuant].skew()
forma_dist = []
for col in vars_cuant:
  skew = asimetria[col]
  if skew > 0.5:
    forma = "Asimétrica positiva (cola a la derecha)"
  elif skew < -0.5:
    forma = "Asimétrica negativa (cola a la izquierda)"
  else:
    forma = "Simétrica"
  forma_dist.append({
      'Variable': col,
      'Forma': forma,
      'Skew': skew
  })

forma_df = pd.DataFrame(forma_dist)
print("\n Forma de la distribución")
display(forma_df)

# Histogramas
for col in vars_cuant:
  if col not in excluir:
    plt.figure()
    df[col].hist()
    plt.title(f'Histograma de {col}')
    plt.xlabel(col)
    plt.ylabel('Frecuencia')
    plt.grid()
    plt.show()
```

```python
# Análisis de correlación  de los datos, mapa de calor
import seaborn as sns

# Correlación
correlacion = df[vars_cuant].corr()

## Mapa de calor
plt.figure(figsize = (10,8))
sns.heatmap(correlacion, annot = True, cmap = 'coolwarm')
plt.title('Mapa de Calor')
plt.show()
```

```python
# Calidad de los datos
# a) Detecta valores faltantes, erróneos, duplicados, errores de escritura
df[variables].isnull().sum() #faltantes
(df[variables].isnull().sum()/len(df))*100


df[variables].duplicated().sum() #duplicados
df[variables][df[variables].duplicated()]
```
